```python
# skitlearn ile model kurmak
# sadece katsayÄ±lar
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(X, Y)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
# katsayÄ±lar parametre tahmin yÃ¶ntemleri farklÄ± olmsaÄ±ndan farklÄ±laÅŸtÄ±
```

**model**

```python
print ("sales = %.2f + TV*%.2f + radio*%.2f + newspaper*%.5f" % (model.intercept_, model.coef_[0][0], model.coef_[0][1], model.coef_[0][2]) )
```

```python
new_data = [
    [30, 10, 40],
    [5, 25, 80],
    [45, 100, 12],
]

new_data = pd.DataFrame(new_data, columns=["TV", "radio", "newspaper"])
new_data
```

```python
model.predict(new_data)
```

```python
for i in range(len(new_data)):
  sales = model.intercept_ + new_data.iloc[i,0]*model.coef_[0][0] + new_data.iloc[i,1]*model.coef_[0][1] + new_data.iloc[i,2]*model.coef_[0][2]
  print(sales)
```

```python
# model performansÄ±
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse = mean_squared_error(Y, model.predict(X))
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
# sÄ±nama seti, holdout
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=99)
```

```python
x_train.head()
```

```python
X.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model2 = LinearRegression().fit(x_train, y_train)
```

```python
# train error
np.sqrt(mean_squared_error(y_train, model2.predict(x_train)))
```

```python
# test error
np.sqrt(mean_squared_error(y_test, model2.predict(x_test)))
```

```python
# k-katlÄ± cross validation
from sklearn.model_selection import cross_val_score
errors = cross_val_score(model2, x_train, y_train, cv=10, scoring="neg_mean_squared_error")
errors
```

```python
# train seti iÃ§inden 10 farklÄ± hata elde edildi en son ortalama alÄ±ndÄ±ÄŸÄ±nda train setinin ortalama hatasÄ± elde ediliyor
```

```python
np.mean(-errors) # cv mse
```

```python
np.sqrt(np.mean(-errors)) # cv rmse
```
